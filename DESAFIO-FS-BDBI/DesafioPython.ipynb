```python
df2 = df.groupby('Department').size().sort_values(ascending=False)
df2.plot(kind='bar')
plt.xlabel('Departamento')
plt.ylabel('Quantidade de funcionários')
plt.title('Quantidade de funcionários por departamento')
plt.show()
```

Comando que plota um gráfico de barras sobre a quantidade de funcionários por departamento. Utilizei os comandos para ajustar o gráfico em ordem decrescente e adicionei título e legenda para facilitar a visualização.

Insight: Mostra que a empresa tem um grande volume de funcionários no departamento de "Research Development"

```python
education_counts = df.groupby('Education').size()

plt.plot(education_counts.index, education_counts.values, marker='o', linestyle='-')
plt.xlabel("Nivel de educação")
plt.ylabel("Quantidade de pessoas")
plt.title("Nivel de distribuição da educação")
plt.show()
```

Criei uma variável para armazenar o agrupamento da coluna 'Education'. Em seguida, plotei o gráfico de linhas com legendas nos eixos X e Y e, por fim, adicionei um título.

Insight: Há uma grande quantidade de pessoas na empresa com o nível de educação 3, ou seja, a maioria dos funcionários possui um nível de educação mediano.

```python
atrito = df['Attrition'].value_counts()

plt.pie(atrito, labels = atrito.index, autopct = '%1.1f%%')
plt.title(" Atrito no trabalho")
plt.axis("equal")
plt.show()
```

Criei uma variável para armazenar o agrupamento da coluna 'Attrition'. Em seguida, plotei o gráfico de pizza.

Insight: Há uma grande quantidade de pessoas na empresa que não tem atrito com outro integrantes na empresa.
